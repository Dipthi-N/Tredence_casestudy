```python
# (Colab usually has these, but safe)
!pip install torch torchvision matplotlib -q

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

trainset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=256, shuffle=True)
testloader = DataLoader(testset, batch_size=256, shuffle=False)
```

```python
class PrunableLinear(nn.Module):
    def __init__(self, in_f, out_f):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(out_f, in_f)*0.01)
        self.bias = nn.Parameter(torch.zeros(out_f))

        # KEY FIX
        self.gate_scores = nn.Parameter(torch.full((out_f, in_f), -2.0))

    def forward(self, x):
        gates = torch.sigmoid(self.gate_scores)
        return F.linear(x, self.weight * gates, self.bias)

    def gates(self):
        return torch.sigmoid(self.gate_scores /0.5)
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = PrunableLinear(3072, 512)
        self.fc2 = PrunableLinear(512, 256)
        self.fc3 = PrunableLinear(256, 10)

    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)

    # 🔥 stronger sparsity push
    def sparsity_loss(self):
        loss = 0
        for l in [self.fc1, self.fc2, self.fc3]:
            loss += torch.mean(l.gates())
        return loss

    def sparsity(self):
        g = torch.cat([l.gates().view(-1) for l in [self.fc1,self.fc2,self.fc3]])
        return (g < 0.05).float().mean().item()*100   # relaxed threshold
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914,0.4822,0.4465),
                         (0.2470,0.2435,0.2616))
])

train = datasets.CIFAR10('./data', train=True, download=True, transform=transform)
test  = datasets.CIFAR10('./data', train=False, download=True, transform=transform)

trainloader = DataLoader(train, batch_size=256, shuffle=True)
testloader  = DataLoader(test, batch_size=256)
```

```python
def test(model):
    model.eval()
    c,t=0,0
    with torch.no_grad():
        for x,y in testloader:
            x,y = x.to(device), y.to(device)
            p = model(x).argmax(1)
            c += (p==y).sum().item()
            t += y.size(0)
    return 100*c/t


def train(model, lam):
    opt = optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(15):
        model.train()
        for x,y in trainloader:
            x,y = x.to(device), y.to(device)

            out = model(x)
            loss = F.cross_entropy(out,y) + lam*model.sparsity_loss()

            opt.zero_grad()
            loss.backward()
            opt.step()

        acc = test(model)
        sp = model.sparsity()
        print(f"E{epoch+1} Acc={acc:.2f}% Sparsity={sp:.2f}%")
```

```python
lambdas = [0.005, 0.01, 0.02]   # strong values

for lam in lambdas:
    print(f"\n===== Lambda = {lam} =====")
    model = Net().to(device)
    train(model, lam)
```

```python
import matplotlib.pyplot as plt

g = torch.cat([model.fc1.gates().view(-1),
               model.fc2.gates().view(-1),
               model.fc3.gates().view(-1)]).detach().cpu().numpy()

plt.hist(g, bins=50)
plt.title("Gate Distribution")
plt.xlabel("Gate Value")
plt.ylabel("Count")
plt.show()
```
